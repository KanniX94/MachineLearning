```python
foodjson = z.open("foods-2011-10-03.json")
foodpd = pd.read_json(foodjson)
```

```python
foodpd
```

Zwischenspeichern der Spalten <I>'description', 'group', 'id'</I> und <I>'manufacturer'</I>

```python
d = {'description': foodpd['description'], 'group': foodpd['group'], 'id': foodpd['id'], 'manufacturer': foodpd['manufacturer']}
info = pd.DataFrame(data=d)
#info
```

```python
foodpd['nutrients'][1]
```

Speichern aller Nährstofflisten in ein Array von DataFrames

```python
dflist = []
for i in range(np.shape(foodpd)[0]):
    value = []
    units = []
    description = []
    group = []
    for j in range(np.shape(foodpd['nutrients'][i])[0]):
        value.append(foodpd['nutrients'][i][j]['value'])
        units.append(foodpd['nutrients'][i][j]['units'])
        description.append(foodpd['nutrients'][i][j]['description'])
        group.append(foodpd['nutrients'][i][j]['group'])
    d = {'id': foodpd['id'][i], 'value': value, 'units': units, 'description': description, 'group': group}
    df = pd.DataFrame(data=d)
    dflist.append(df)
```

```python
#foodpd['nutrients'][0][0]['group']
```

Vereinigung des Arrays mithilfe der Funktion <I>pandas.concat()</I> zu einem großen DataFrame
<I>nutrients</I> (389355 Einträge)

```python
nutrients = pd.concat(dflist)
```

```python
nutrients
```

```python
nutrients = nutrients.rename(columns = {"description": "n_description", 
                                  "group":"n_group"}) 
```

```python
combine = pd.merge(nutrients, info, on='id',
how='outer')
combine.head()
```

Alle Einträge, die Zink enthalten rausschreiben

```python
Zink = combine.loc[combine['n_description'] == 'Zinc, Zn']
Zink.head()
```

```python
Zink.hist(column='value', figsize=(15,8), bins=50)
```
